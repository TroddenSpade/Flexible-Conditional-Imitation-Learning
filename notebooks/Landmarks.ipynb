```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


DATA_PATH = './Airsim'
```

```python
folders = os.listdir(DATA_PATH)
```

```python
def plot(sets, colors=None):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    for i, s in enumerate(sets):
        if colors:
            ax.scatter(s[:,0], s[:,1], color=colors[i])
        else:
            ax.scatter(s[:,0], s[:,1])
    ax.set_xlim(0, 140)
    ax.set_ylim(70, -70)
    plt.show()
```

```python
path = os.path.join(DATA_PATH, folders[0], 'airsim_rec.txt')
samples1 = pd.read_csv(path, sep='\t')
samples1.head()
```

```python
X1 = samples1["POS_X"].to_numpy()
Y1 = samples1["POS_Y"].to_numpy()
Z1 = samples1["POS_Z"].to_numpy()
```

```python
# route_set = np.vstack((X1,Y1)).T
x_r = np.expand_dims(X1, axis=-1)
y_r = np.expand_dims(Y1, axis=-1)
route_set = np.hstack((x_r,y_r))
```

```python
path = os.path.join(DATA_PATH, folders[7], 'airsim_rec.txt')
samples = pd.read_csv(path, sep='\t')

x_s = np.expand_dims(samples["POS_X"].to_numpy(), axis=-1)
y_s = np.expand_dims(samples["POS_Y"].to_numpy(), axis=-1)
sample_set = np.hstack((x_s,y_s))
```

```python
plot([sample_set, route_set])
```

```python
sample_set.shape
```

```python
idx = [10, 60, 80, 100]

plot([route_set, sample_set[idx]], ['g','r'])
```

```python
def nearest(point, set_of_points):
    diff = np.square(set_of_points - point)
    diff = diff.sum(axis=1)
    return np.argmin(diff)
```

```python
array = []
for id in idx:
    n_id = nearest(sample_set[id], route_set)
    array.append(route_set[n_id:n_id+10])
array.append(sample_set[idx])
plot(array)
```
